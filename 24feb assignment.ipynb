<b> Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row. <br> Ans :</b>
```
import pandas as pd
def reindex(dataset):
    newindex=range(1,2*len(dataset),2)
    return dataset.set_index(pd.Index(newindex)) 
```

    df["Moving Average"]=df["sales"].rolling(window=7,min_periods=1).mean()
